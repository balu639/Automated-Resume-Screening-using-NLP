```python
from transformers import pipeline
# Define the pipeline with the BART model
pipe = pipeline('summarization', model='facebook/bart-large-cnn', tokenizer='facebook/bart-large-cnn')
# pipe = pipeline('summarization', model='google/pegasus-cnn_dailymail')

def generate_summary(input_text):
  # Split the input text into smaller segments
  segment_length = 2000  # You can adjust this value as needed
  segments = [input_text[i:i+segment_length] for i in range(0, len(input_text), segment_length)]
  # Summarize each segment separately
  summaries = []
  for segment in segments:
      summary = pipe(segment, max_length=150)  # You can adjust max_length as needed
      summaries.append(summary[0]['summary_text'])

  # Concatenate the summarized segments to get the complete summary
  complete_summary = ' '.join(summaries)
  return complete_summary
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def preprocess_text(text):
    # You can add your preprocessing steps here, such as lowercasing, removing stopwords, etc.
    return text

def calculate_tfidf(text1, text2):
    # Initialize TF-IDF vectorizer
    vectorizer = TfidfVectorizer()

    # Fit the vectorizer on the combined texts
    corpus = [text1, text2]
    tfidf_matrix = vectorizer.fit_transform(corpus)

    # Get TF-IDF vectors for both texts
    tfidf_text1 = tfidf_matrix[0]
    tfidf_text2 = tfidf_matrix[1]

    return tfidf_text1, tfidf_text2, vectorizer.get_feature_names_out()

def select_relevant_features(tfidf_text1, tfidf_text2, feature_names):
    # Calculate cosine similarity between TF-IDF vectors to find relevant features
    similarity_score = cosine_similarity(tfidf_text1, tfidf_text2)

    # Select features with highest similarity scores
    relevant_feature_indices = similarity_score.argsort(axis=1)[:, ::-1][:, :10]  # Select top 10 features
    relevant_features = [feature_names[i] for i in relevant_feature_indices[0]]

    return relevant_features

def calculate_similarity(tfidf_text1, tfidf_text2):
    # Calculate cosine similarity between TF-IDF weighted vectors
    similarity_score = cosine_similarity(tfidf_text1, tfidf_text2)
    return similarity_score[0][0]
```

```python
def finding_similarity(text1, text2):
  # Calculate TF-IDF for both texts
  tfidf_text1, tfidf_text2, feature_names = calculate_tfidf(text1, text2)
  # Select relevant features using TF-IDF
  relevant_features = select_relevant_features(tfidf_text1, tfidf_text2, feature_names)
  # Calculate similarity between TF-IDF weighted vectors
  similarity_score = calculate_similarity(tfidf_text1, tfidf_text2)
  return similarity_score
```

# Finetuning the BART Model

```python
import pandas as pd
from torch.utils.data import Dataset, DataLoader
from transformers import BartTokenizer, BartForConditionalGeneration, AdamW, get_scheduler
from sklearn.model_selection import train_test_split
import torch

# Load the dataset
df = pd.read_csv('/content/dataset/resume_and_summaries.csv')
texts = df['Resume'].tolist()
summaries = df['Resume_Summarized'].tolist()

# Split the data into training and validation sets
texts_train, texts_val, summaries_train, summaries_val = train_test_split(texts, summaries, test_size=0.1, random_state=42)

# Define the dataset class
class ResumeDataset(Dataset):
    def __init__(self, tokenizer, texts, summaries, max_length=512):
        self.tokenizer = tokenizer
        self.texts = texts
        self.summaries = summaries
        self.max_length = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, index):
        source_encodings = self.tokenizer(self.texts[index], truncation=True, padding='max_length', max_length=self.max_length)
        target_encodings = self.tokenizer(self.summaries[index], truncation=True, padding='max_length', max_length=self.max_length)

        return {"input_ids": torch.tensor(source_encodings['input_ids']),
                "attention_mask": torch.tensor(source_encodings['attention_mask']),
                "labels": torch.tensor(target_encodings['input_ids'])}

# Load tokenizer
tokenizer = BartTokenizer.from_pretrained('facebook/bart-large-cnn')

# Prepare datasets
train_dataset = ResumeDataset(tokenizer, texts_train, summaries_train)
val_dataset = ResumeDataset(tokenizer, texts_val, summaries_val)

train_loader = DataLoader(train_dataset, batch_size=2, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=2, shuffle=False)
```

```python
import torch
from transformers import BartTokenizer, BartForConditionalGeneration, AdamW, get_scheduler
from tqdm import tqdm
import numpy as np

# Assuming model and tokenizer are already loaded and training/validation DataLoaders are set up

# Load the tokenizer and model
model = BartForConditionalGeneration.from_pretrained('facebook/bart-large-cnn')
tokenizer = BartTokenizer.from_pretrained('facebook/bart-large-cnn')
model.to(torch.device('cuda' if torch.cuda.is_available() else 'cpu'))

# Prepare optimizer and scheduler
optimizer = AdamW(model.parameters(), lr=5e-5)
num_training_steps = len(train_loader) * 3  # for 3 epochs
lr_scheduler = get_scheduler("linear", optimizer=optimizer, num_warmup_steps=0, num_training_steps=num_training_steps)

# Function to calculate accuracy
def calculate_accuracy(outputs, labels):
    predictions = torch.argmax(outputs.logits, dim=-1)
    correct = (predictions == labels) & (labels != -100)  # ignore padding for accuracy calculation
    return torch.sum(correct).item() / torch.sum(labels != -100).item()

# Training loop with tqdm
model.train()
for epoch in range(10):  # train for 3 epochs
    progress_bar = tqdm(train_loader, desc=f"Epoch {epoch + 1}")
    total_loss = 0
    total_accuracy = 0
    for batch in progress_bar:
        batch = {k: v.to(model.device) for k, v in batch.items()}
        outputs = model(**batch)
        loss = outputs.loss

        # Calculate "accuracy"
        accuracy = calculate_accuracy(outputs, batch['labels'])

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        lr_scheduler.step()

        total_loss += loss.item()
        total_accuracy += accuracy
        progress_bar.set_postfix(loss=loss.item(), accuracy=accuracy)

    avg_loss = total_loss / len(train_loader)
    avg_accuracy = total_accuracy / len(train_loader)
    print(f"Average Loss: {avg_loss:.2f}")
    print(f"Average Accuracy: {avg_accuracy:.2f}")

# Save the fine-tuned model
model.save_pretrained('model/fine_tuned_bart1')
tokenizer.save_pretrained('model/fine_tuned_bart1')
```

```python
def generate_summary(text, model, tokenizer, max_length=512, num_beams=4):
    model.eval()  # Set the model to evaluation mode
    with torch.no_grad():
      inputs = tokenizer.encode("summarize: " + text, return_tensors="pt", max_length=max_length, truncation=True).to(model.device)
      summary_ids = model.generate(inputs, max_length=150, min_length=40, length_penalty=2.0, num_beams=num_beams, early_stopping=True)
      summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary
```

```python
from rouge_score import rouge_scorer

# Generate summaries
generated_summaries = [generate_summary(text, model, tokenizer) for text in texts_val]

# Initialize ROUGE scorer
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
scores = {key: [] for key in ['rouge1', 'rouge2', 'rougeL']}

# Calculate ROUGE scores
for ref, gen in zip(summaries_val, generated_summaries):
    score = scorer.score(ref, gen)
    for key in scores.keys():
        scores[key].append(score[key].fmeasure)

# Compute average ROUGE scores
average_scores = {key: sum(val)/len(val) for key, val in scores.items()}
```

```python
import torch
from transformers import BertModel, BertTokenizer

def explain_bert_embeddings(texts):
    # Load pre-trained model tokenizer (vocabulary)
    tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

    # Load pre-trained model
    model = BertModel.from_pretrained('bert-base-uncased')
    model.eval()  # Set the model to evaluation mode

    # Device configuration
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    # Tokenizing text
    encoded_input = tokenizer(texts, padding=True, truncation=True, return_tensors='pt')
    input_ids = encoded_input['input_ids'].to(device)
    attention_mask = encoded_input['attention_mask'].to(device)

    # Compute token embeddings
    with torch.no_grad():
        outputs = model(input_ids, attention_mask=attention_mask)

    # Get the embeddings of the last layer
    embeddings = outputs.last_hidden_state.mean(dim=1)

    # Explain embeddings calculation
    print("\nSentence Embeddings (mean of last layer hidden states):")
    print(embeddings)

    return embeddings

embeddings = explain_bert_embeddings(merged_df1['Resume_str'][1])
```
